```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.cluster import KMeans
```

```python
df=pd.read_csv('Iris.csv')
df.head()
```

```python
df.set_index('Id',drop=True,inplace=True)
```

```python
df.info()
```

```python
sns.set_style('whitegrid')
sns.lmplot(x='SepalLengthCm',y='SepalWidthCm',hue='Species',data=df,fit_reg=False)
```

```python
sns.lmplot(x='PetalLengthCm',y='PetalWidthCm',hue='Species',data=df,fit_reg=False)
```

```python
X=df.drop('Species',axis=1)
y=df['Species']
```

```python
clust=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,random_state=101)
    kmeans.fit(X)
    clust.append(kmeans.inertia_)
plt.plot(range(1,11),clust)
```

```python
model=KMeans(n_clusters=3)
```

```python
predictions=model.fit_predict(X)
```

```python
def clustering(species):
    if species=='Iris-setosa':
        return 1
    if species=='Iris-versicolor':
        return 2
    else:
        return 0
    
```

```python
df['species']=df['Species'].apply(clustering)
```

```python
from sklearn.metrics import classification_report
```

```python
def accuracy(y_test,predictions):
    acc=0
    for i in range(len(predictions)):
        if y_test[i+1]==predictions[i]:
            acc+=1
        else:
            acc+=0
    return acc/len(predictions)
```

```python
accuracy(df['species'],predictions)
```

```python
plt.scatter(X.values[predictions==0,0],X.values[predictions==0,1],color='red',label='Iris-virginica')
plt.scatter(X.values[predictions==1,0],X.values[predictions==1,1],color='blue',label='Iris-setosa')
plt.scatter(X.values[predictions==2,0],X.values[predictions==2,1],color='green',label='Iris-versicolour')
plt.scatter(model.cluster_centers_[:, 0],model.cluster_centers_[:,1],c = 'yellow', label = 'Centroids')
plt.legend()
```

```python
plt.scatter(X.values[df['species'].values==0,0],X.values[df['species'].values==0,1],color='red',label='Iris-virginica')
plt.scatter(X.values[df['species'].values==1,0],X.values[df['species'].values==1,1],color='blue',label='Iris-setosa')
plt.scatter(X.values[df['species'].values==2,0],X.values[df['species'].values==2,1],color='green',label='Iris-versicolour')
plt.legend()
```
